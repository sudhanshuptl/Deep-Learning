```python
from tensorflow import keras
import numpy as np
```

```python
xs = np.array([-1,0,1,2,3,4], dtype=float)
ys = np.array([-3,-1,1,3,5,7], dtype=float)
```

```python
# create a deep layer with 1 node
# nomber of data will be inputed to this layer is 1 (input shape)
# dense :  Stands to define layer of connected neurons 
model  = keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1]) # we can add next layer after adding , 
])
```

```python
# Compile model 
model.compile(optimizer='sgd', loss='mean_squared_error')
# loss :  we selected a function to calculate the loss after prediction 
# then Optimiser sgd (gradient decent) will optimise neurons with learning rate to minimise loss
```

```python
# print(model.compile.__doc__)  #print docstring
```

```python
model.fit(xs, ys, epochs=500) # epoches : number of time it trains for all data
```

```python
model.predict([10.0]) # answer should be 19
# but it return value close to 19 but not exact 19 
```
